```python
import numpy as np
import datetime as dt
```

```python
import pandas as pd
df = pd.read_csv(r'Online Sales Data.csv')
df
```

```python
#droping the duplicates from the data if present
df = df.drop_duplicates()
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.strftime('%B')
df
```

```python
df.info()
```

```python
#sorting the data on the bases of product category
df.sort_values(by="Product Category", ascending=True)
```

```python
#Total revenue generated by all the products
df['Total Revenue'].sum()
```

```python
#Total revenue based on product category
df2 = df.groupby("Product Category").agg({'Total Revenue' : 'sum'}).reset_index()
df2
```

```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import plotly.express as px
```

```python
#Graph of total revenue based on product category
fig = px.bar(df2, x='Product Category', y='Total Revenue',
                  color='Product Category',
             title='Total Revenue based on product category')
fig.show()
```

```python
#Total revenue generated based on region
dz = df.groupby("Region").agg({"Total Revenue": "sum"}).reset_index()
dz
```

```python
#Graph of total revenue generated based on region
fig = px.bar(dz, x='Region', y='Total Revenue', color='Region',
             title='Total Revenue based on Region')
fig.show()
```

```python
#Top 10 highest revenue generating products
dz2 = df.groupby('Product Name').agg({"Total Revenue": "sum"}).reset_index()
dz2 = df.sort_values(by='Total Revenue', ascending=False).head(10)
dz2
```

```python
#Graphical Represenatation of highest revenue genarating products
fig = px.pie(dz2, values='Total Revenue',
              names='Product Name', title='Top 10 products with high revenue')
fig.show()
```

```python
#Lowest revenue genarating product
dz2 = df.groupby('Product Name').agg({"Total Revenue": "sum"}).reset_index()
dz2 = df.sort_values(by='Total Revenue', ascending=True).head(10)
dz2
```

```python
#Graphical Representation of lowest revenue generating products
fig = px.pie(dz2, values='Total Revenue',
              names='Product Name', title='Lowest Revenue generating products')
fig.show()
```

```python
#Total number of units sold based on product category
df3 = df.groupby("Product Category").agg({"Units Sold": "sum"}).reset_index()
df3
```

```python
#Graphical represenatation of total number of units sold based on product category
fig = px.bar(df3, x='Product Category', y='Units Sold', color='Product Category',
             title='Total number of Units Sold based on product category')
fig.show()
```

```python
#Total number of units sold every month
df4 = df.groupby("month").agg({"Units Sold": "sum"}).reset_index()
df4 = df4.sort_values(by='Units Sold', ascending=False)
df4
```

```python
#Graphical representation of total number of units sold every month
fig = px.bar(df4, x='month', y='Units Sold', color='month',
             title='Total number of Units Sold based every month')
fig.show()
```

```python
#Total Revenue in every month
dz3 = df.groupby("month").agg({"Total Revenue": "sum"}).reset_index()
dz3 = dz3.sort_values(by='Total Revenue', ascending=False)
dz3
```

```python
#Bar graph of total revenue gained in every month
fig = px.bar(dz3, x='month', y='Total Revenue', color='month',
             title='Total Revenue gained every month')
fig.show()
```

```python
#Total revenue generated by month and product category
df5 = df.groupby(['month', 'Product Category'])[['Units Sold', 'Total Revenue']].sum().reset_index()
df5
```

```python
#Total revenue generated in every month by units sold
df6 = df.groupby(['month'])[['Units Sold', 'Total Revenue']].sum().reset_index()
df6
```

```python
#Total revenue generated by region and units sold
df7 = df.groupby(['Region'])[['Units Sold', 'Total Revenue']].sum().reset_index()
df7
```

```python
#Total revenue generated by region and payment method
df8 = df.groupby(['Region', 'Payment Method'])[['Total Revenue']].sum().reset_index()
df8
```
